```python
MISSING = {"--", "n/a", "unknown", "", "nan", "none"}

def parse_revenue(s):
    if pd.isna(s): return np.nan
    s = str(s).strip().lower().replace(",","").replace("$","").replace("usd","").strip()
    if s in MISSING: return np.nan
    mult = 1
    if "billion" in s:
        mult = 1e9; s = re.sub(r"billion","",s).strip()
    elif "million" in s or re.search(r"\d\s*mn$",s) or re.search(r"\d\s*m$",s):
        mult = 1e6; s = re.sub(r"(million|\bmn\b|\bm\b)","",s).strip()
    m = re.search(r"[\d.]+", s)
    return float(m.group()) * mult if m else np.nan

def parse_pct(s):
    if pd.isna(s): return np.nan
    s = str(s).strip().replace("+","").replace("%","").lower()
    if s in MISSING: return np.nan
    try: return float(s)
    except: return np.nan

def parse_rd(s, annual_rev):
    if pd.isna(s): return np.nan
    raw = str(s).strip()
    if raw.lower() in MISSING: return np.nan
    if "% rev" in raw.lower():
        m = re.match(r"[\d.]+", raw)
        if m and not np.isnan(annual_rev): return float(m.group())/100 * annual_rev
        return np.nan
    return parse_revenue(raw)

def parse_team_size(s):
    if pd.isna(s): return np.nan
    s = str(s).strip().lower().replace(",","")
    if s.endswith("k"):
        try: return float(s[:-1]) * 1000
        except: return np.nan
    try: return float(s)
    except: return np.nan

def standardize_ai(s):
    if pd.isna(s): return np.nan
    val = str(s).strip().lower()
    if val in {"yes","adopted","ai enabled","live","production"}: return 1
    if val in {"no","not yet","legacy only","manual only"}:      return 0
    return np.nan  

```


## Step 5: Clean: Apply Transformations & Drop Missing

```python
df["ANNUAL_REV_NUM"]    = df["ANNUAL_REV"].apply(parse_revenue)
df["REV_GROWTH_NUM"]    = df["REV_GROWTH"].apply(parse_pct)
df["RD_SPEND_NUM"]      = df.apply(lambda r: parse_rd(r["RD_SPEND"], r["ANNUAL_REV_NUM"]), axis=1)
df["TEAM_SIZE_NUM"]     = df["TEAM_SIZE"].apply(parse_team_size)
df["DIGITAL_SALES_NUM"] = df["DIGITAL_SALES"].apply(parse_pct)
df["FIRM_AGE"]          = 2026 - pd.to_numeric(df["FOUNDED"], errors="coerce")
df["AI_ADOPTED"]        = df["AI_STATUS"].apply(standardize_ai)

# Drop Missing:
df_clean = df.dropna(subset=["REV_GROWTH_NUM","AI_ADOPTED"]).copy()
df_clean["AI_ADOPTED"] = df_clean["AI_ADOPTED"].astype(int)
df_clean.reset_index(drop=True, inplace=True)
df_clean.to_csv("clean_data.csv", index=False)

print(f"✓ Clean dataset ready")
print(f"  Total firms     : {len(df_clean)}")
print(f"  AI = 1 (adopted): {df_clean['AI_ADOPTED'].sum()}")
print(f"  AI = 0 (not)    : {(df_clean['AI_ADOPTED']==0).sum()}")
df_clean[["FIRM","SEGMENT","AI_ADOPTED","REV_GROWTH_NUM","ANNUAL_REV_NUM"]].head(5)
```


## Step 6: Analyze: Naive OLS (Baseline, Biased)

```python
def ols_fit(y, X):
    n, k = X.shape
    b, _, _, _ = np.linalg.lstsq(X, y, rcond=None)
    resid  = y - X @ b
    sigma2 = (resid @ resid) / (n - k)
    cov    = sigma2 * np.linalg.inv(X.T @ X)
    se     = np.sqrt(np.diag(cov))
    t      = b / se
    p      = 2 * (1 - stats.t.cdf(np.abs(t), df=n-k))
    r2     = 1 - (resid @ resid) / ((y - y.mean())**2).sum()
    return b, se, t, p, r2

y_all    = df_clean["REV_GROWTH_NUM"].values
X_naive  = np.column_stack([np.ones(len(df_clean)), df_clean["AI_ADOPTED"].values])

b0, se0, t0, p0, r20 = ols_fit(y_all, X_naive)
naive_coef = b0[1]
naive_pval = p0[1]

print("=" * 42)
print("       NAIVE OLS RESULTS")
print("=" * 42)
print(f"  Intercept  :  {b0[0]:+.4f}   p = {p0[0]:.4f}")
print(f"  AI_ADOPTED :  {b0[1]:+.4f}   p = {p0[1]:.4f}")
print(f"  R-squared  :  {r20:.4f}")
print("=" * 42)
print(f"\n→ Naive estimate: AI adoption associated with +{naive_coef:.2f} pp revenue growth")
```


## Step 7: Analyze: Estimate Propensity Scores


```python
covariates   = ["ANNUAL_REV_NUM","FIRM_AGE","TEAM_SIZE_NUM","DIGITAL_SALES_NUM"]
df_psm       = df_clean[["FIRM","AI_ADOPTED","REV_GROWTH_NUM"] + covariates].dropna().copy()
df_psm.reset_index(drop=True, inplace=True)

df_psm["LOG_REV"]  = np.log(df_psm["ANNUAL_REV_NUM"])
psm_features       = ["LOG_REV","FIRM_AGE","TEAM_SIZE_NUM","DIGITAL_SALES_NUM"]

scaler    = StandardScaler()
X_scaled  = scaler.fit_transform(df_psm[psm_features].values)
y_ps      = df_psm["AI_ADOPTED"].values

logit = LogisticRegression(max_iter=1000, C=1.0, solver="lbfgs")
logit.fit(X_scaled, y_ps)
df_psm["PROP_SCORE"] = logit.predict_proba(X_scaled)[:, 1]


print(f"  n = {len(df_psm)}  |  Treated (AI=1) = {y_ps.sum()}  |  Control (AI=0) = {(y_ps==0).sum()}")
df_psm[["FIRM","AI_ADOPTED","PROP_SCORE"]].head(5)
```


## Step 8: Analyze: Common Support Plot


```python
treated_ps = df_psm[df_psm["AI_ADOPTED"] == 1]["PROP_SCORE"]
control_ps = df_psm[df_psm["AI_ADOPTED"] == 0]["PROP_SCORE"]

fig, ax = plt.subplots(figsize=(8, 4.5))
ax.hist(control_ps, bins=14, alpha=0.55, color="#1f618d", label="Not Adopted (AI=0)", density=True)
ax.hist(treated_ps,  bins=14, alpha=0.55, color="#c0392b", label="Adopted (AI=1)",     density=True)
ax.set_xlabel("Propensity Score", fontsize=12)
ax.set_ylabel("Density", fontsize=12)
ax.set_title("Common Support: Propensity Score Distributions", fontsize=13, fontweight="bold")
ax.legend(fontsize=11)
plt.tight_layout()
plt.savefig("fig_common_support.png", dpi=150)
plt.show()
```


##  Step 9:Analyze: SMD Before Matching


```python
FEATURE_LABELS = {
    "LOG_REV":           "Log Annual Revenue",
    "FIRM_AGE":          "Firm Age (years)",
    "TEAM_SIZE_NUM":     "Team Size",
    "DIGITAL_SALES_NUM": "Digital Sales (%)"
}

def compute_smd(df, features, treat_col="AI_ADOPTED"):
    t = df[df[treat_col] == 1]
    c = df[df[treat_col] == 0]
    out = {}
    for f in features:
        m1, m0 = t[f].mean(), c[f].mean()
        s1, s0 = t[f].std(ddof=1), c[f].std(ddof=1)
        pooled = np.sqrt((s1**2 + s0**2) / 2)
        out[f] = round((m1 - m0) / pooled, 4) if pooled > 0 else np.nan
    return out

smd_before = compute_smd(df_psm, psm_features)

print("SMD Before Matching  (|SMD| > 0.1 = imbalanced):")
print("-" * 48)
for k, v in smd_before.items():
    flag = " IMBALANCED" if abs(v) > 0.1 else "✓"
    print(f"  {FEATURE_LABELS[k]:25s}: {v:+.4f}  {flag}")
```


## Step 10: Analyze: 1-to-1 Nearest-Neighbor PSM Matching


```python
treated_df = df_psm[df_psm["AI_ADOPTED"] == 1].copy().reset_index(drop=True)
control_df = df_psm[df_psm["AI_ADOPTED"] == 0].copy().reset_index(drop=True)

nn = NearestNeighbors(n_neighbors=1, algorithm="ball_tree")
nn.fit(control_df[["PROP_SCORE"]].values)
_, indices = nn.kneighbors(treated_df[["PROP_SCORE"]].values)

matched_control = control_df.iloc[indices.flatten()].copy()
df_matched      = pd.concat([treated_df, matched_control], ignore_index=True)

print(f"  {len(treated_df)} treated  +  {len(matched_control)} matched controls  =  {len(df_matched)} total rows")
```


## Step 11: Analyze: SMD After Matching + Love Plot


```python
smd_after = compute_smd(df_matched, psm_features)

print("SMD After Matching:")
print("-" * 48)
for k, v in smd_after.items():
    flag = " still imbalanced" if abs(v) > 0.1 else "✓  balanced"
    print(f"  {FEATURE_LABELS[k]:25s}: {v:+.4f}  {flag}")

# Love Plot
fig, ax = plt.subplots(figsize=(8, 4.5))
y_pos  = np.arange(len(psm_features))
b_vals = [abs(smd_before[f]) for f in psm_features]
a_vals = [abs(smd_after[f])  for f in psm_features]

ax.scatter(b_vals, y_pos, color="#c0392b", s=90, label="Before Matching", zorder=3)
ax.scatter(a_vals, y_pos, color="#1f618d", s=90, marker="D", label="After Matching", zorder=3)
ax.axvline(x=0.1, color="gray", linestyle="--", linewidth=1.2, label="SMD = 0.1 threshold")
ax.set_yticks(y_pos)
ax.set_yticklabels([FEATURE_LABELS[f] for f in psm_features], fontsize=11)
ax.set_xlabel("|Standardized Mean Difference|", fontsize=11)
ax.set_title("Love Plot: Covariate Balance Before and After PSM", fontsize=12, fontweight="bold")
ax.legend(fontsize=10)
plt.tight_layout()
plt.savefig("fig_love_plot.png", dpi=150)
plt.show()
```


## Step 12: Analyze: Post-Matching OLS (True Causal Effect)


```python
y_m = df_matched["REV_GROWTH_NUM"].values
X_m = np.column_stack([np.ones(len(df_matched)), df_matched["AI_ADOPTED"].values])

bm, sem, tm, pm, r2m = ols_fit(y_m, X_m)
psm_coef = bm[1]
psm_pval = pm[1]

print("=" * 42)
print("    POST-MATCHING OLS RESULTS")
print("=" * 42)
print(f"  Intercept  :  {bm[0]:+.4f}   p = {pm[0]:.4f}")
print(f"  AI_ADOPTED :  {bm[1]:+.4f}   p = {pm[1]:.4f}")
print(f"  R-squared  :  {r2m:.4f}")
print("=" * 42)
```


## Step 13: Analyze: Final Comparison + Robustness Check

```python

# Welch's t-test on matched sample
g1 = df_matched[df_matched["AI_ADOPTED"] == 1]["REV_GROWTH_NUM"]
g0 = df_matched[df_matched["AI_ADOPTED"] == 0]["REV_GROWTH_NUM"]
t_stat, t_pval = stats.ttest_ind(g1, g0, equal_var=False)

print("=" * 50)
print("           FINAL COMPARISON SUMMARY")
print("=" * 50)
print(f"  Naive OLS coefficient    :  {naive_coef:+.4f} pp  (p={naive_pval:.4f})")
print(f"  PSM-Adjusted coefficient :  {psm_coef:+.4f} pp  (p={psm_pval:.4f})")
print(f"  Selection-bias removed   :  {naive_coef - psm_coef:+.4f} pp")
print("=" * 50)
print(f"\nRobustness t-test: mean diff={g1.mean()-g0.mean():.4f} pp | t={t_stat:.4f} | p={t_pval:.4f}")

# Bar chart
fig, ax = plt.subplots(figsize=(6, 4))
bars = ax.bar(["Naive OLS","PSM-Adjusted OLS"],
              [naive_coef, psm_coef],
              color=["#e74c3c","#1f618d"], width=0.4, edgecolor="black")
ax.axhline(0, color="black", linewidth=0.8)
ax.set_ylabel("Coefficient on AI Adoption (pp)", fontsize=11)
ax.set_title("Effect of AI Adoption on Revenue Growth:\nNaive vs PSM-Adjusted", fontsize=11, fontweight="bold")
for bar, c in zip(bars, [naive_coef, psm_coef]):
    ax.text(bar.get_x()+bar.get_width()/2, c+0.05, f"{c:.2f} pp",
            ha="center", fontsize=11, fontweight="bold")
plt.tight_layout()
plt.savefig("fig_coef_comparison.png", dpi=150)
plt.show()
```
